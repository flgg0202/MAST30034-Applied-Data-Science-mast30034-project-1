```python
import pandas as pd
import geopandas as gpd

sf = gpd.read_file("../data/landing/taxi_zones.shp")
zones = pd.read_csv("../data/landing/taxi+_zone_lookup.csv")
sf['geometry'] = sf['geometry'].to_crs("+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs")
sf.head()


gdf = gpd.GeoDataFrame(
    pd.merge(zones, sf, on='LocationID', how='inner')
)
```

```python
joined = pd.read_csv('../data/curated/joined.csv')
test_joined = pd.read_csv('../data/curated/test_joined.csv')
```

```python

result_test = test_joined.groupby('LocationID').size()

test_df_crime = result_test.reset_index(name='crime_count')

test_df_crime.to_csv('../data/curated/test_df_crime.csv')


result = joined.groupby('LocationID').size()

result_df_crime = result.reset_index(name='crime_count')

result_df_crime.to_csv('../data/curated/result_df_crime.csv')
```

```python
final = pd.read_csv('../data/curated/final_pd.csv')
```

```python
merged_df = pd.merge(final, result_df_crime, on='LocationID', how='inner')
merged_df = pd.merge(merged_df, gdf, on='LocationID', how='inner')
merged_df
```

```python
merged_df.to_csv('../data/curated/merged_df.csv')
```
